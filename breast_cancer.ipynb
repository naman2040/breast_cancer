<h3>To predict if the Breast Cancer is 'Benign' Or 'Malignant', using given Dataset.</h3>
To Build a Classification Model to produce Accurate Prediction on the Data Entered.


```python
import numpy as np
import pandas as pd
```

# Loading data

```python
df = pd.read_csv("breast_cancer.csv")
df
```

```python
df["Class"].unique()
```

```python
x=df.iloc[:,1:-1].values
y=df.iloc[:,-1].values.reshape(-1,1)
```

# Splitting Data into Test and Train set

```python
from sklearn.model_selection import train_test_split,cross_val_score
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x_train_s=ss.fit_transform(x_train)
x_test_s=ss.transform(x_test)
```

# Applying Logistic Regression Cassification

```python
from sklearn.linear_model import LogisticRegression
classi=LogisticRegression(random_state=0)
classi.fit(x_train_s,y_train)
y_hat=classi.predict(x_test_s)
```

# Fiding Accuracy on Test Set

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_hat)
```

# K Folds on the test set (k=10)

```python
k_fold_prediction=cross_val_score(estimator=classi,X=x_train_s,y=y_train.ravel(),cv=10)
print("Average Accuracy of our Classification Model is : ",k_fold_prediction.mean()*100)
print("Standard Deviation of our Classification Model is : ",k_fold_prediction.std()*100)
```

## Thus Our Logistic Regression Model has Accuracy of 96.7%

# K Nearest Neighbour

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5,metric="minkowski",p=2)
knn.fit(x_train_s,y_train.ravel())
y_pred=knn.predict(x_test_s)
print(accuracy_score(y_test,y_pred)*100)
```

# Niave Bayes

```python
from sklearn.naive_bayes import GaussianNB
gn = GaussianNB()
gn.fit(x_train_s,y_train.ravel())
y_pred=gn.predict(x_test_s)
accuracy_score(y_test,y_pred)*100
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(random_state=4,criterion="entropy")
dtc.fit(x_train_s,y_train)
y_pred=dtc.predict(x_test_s)
accuracy_score(y_test,y_pred)
```

# Random forest

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100,random_state=42)
rf.fit(x_train_s,y_train.ravel())
y_pred=rf.predict(x_test_s)
accuracy_score(y_test,y_pred)*100
```

# Deep Learning ANN model

```python
import tensorflow as tf
```

```python
y_train[y_train==2]=1
y_train[y_train==4]=0
y_test[y_test==2]=1
y_test[y_test==4]=0
```

```python

ann = tf.keras.models.Sequential()
ann.add(tf.keras.layers.Dense(units=10,activation="relu"))
ann.add(tf.keras.layers.Dense(units=9,activation="relu"))
ann.add(tf.keras.layers.Dense(units=1,activation="sigmoid"))
ann.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
ann.fit(x_train_s,y_train,batch_size=21,epochs=50)
```

```python
y_pred=ann.predict(x_test_s)
y_pred[y_pred<0.5]=0
y_pred[y_pred>=0.5]=1
accuracy_score(y_test,y_pred)*100
```

# XGBoost 

```python
from xgboost import XGBClassifier
xgb = XGBClassifier()
xgb.fit(x_train_s,y_train)
y_pred = xgb.predict(x_test_s)
accuracy_score(y_test,y_pred)*100
```

# catBoost

```python
from catboost import CatBoostClassifier
cb = CatBoostClassifier()
cb.fit(x_train_s,y_train)
y_pred = cb.predict(x_test_s)
accuracy_score(y_test,y_pred)*100
```
